***Step 1: Import Required Libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style for all plots
sns.set_style('whitegrid')
print('Libraries imported successfully!')
```

***Step 2: Load the Dataset***

```python
# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

print(f'Shape of dataset: {df.shape}')  # (rows, columns)
```

***Step 3: Understand the Dataset Structure***

```python
# Show the first 5 rows of the dataset
print('First 5 rows of the dataset:', df.head())
```

```python
# Display column names
print('Column names:', df.columns.tolist())
```

```python
# Show data types of each column
print('\nData types:')
print(df.dtypes)
```

```python
# Check for missing values
print('Missing values in each column:')
print(df.isnull().sum())
```

```python
# Summary statistics — mean, min, max, std deviation, etc.
print('Statistical Summary:',df.describe())
```

```python
# Count of each species in the dataset
print('Count of each species:')
print(df['species'].value_counts())
```

***Step 4: Visualizations***

***4.1 Scatter Plot — Relationship Between Variables***

```python
plt.figure(figsize=(8, 5))

# Scatter plot: sepal_length vs petal_length, colored by species
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species',      # color each species differently
    palette='Set1',
    s=80               # size of dots
)

plt.title('Sepal Length vs Petal Length by Species', fontsize=14)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

Observation:
*   Setosa (red) is clearly separated from the other two species.
*   Versicolor and Virginica overlap slightly but are mostly distinguishable.




***4.2 Histogram — Distribution of Each Feature***

```python
df_melted = df.melt(id_vars='species', var_name='feature', value_name='value')

g = sns.FacetGrid(df_melted, col='feature', hue='species', col_wrap=2, height=3)
g.map(sns.histplot, 'value', bins=10, alpha=0.6)
g.add_legend()
plt.show()
```

Observation:
* Petal length and petal width show the most separation between species.



***4.3 Box Plot — Outliers and Spread of Values***

```python
df_melted = df.melt(id_vars='species', var_name='feature', value_name='value')

g = sns.FacetGrid(df_melted, col='feature', col_wrap=2, height=4)
g.map_dataframe(sns.boxplot, x='species', y='value', palette='Set2')
plt.show()
```

Observation:
*   Sepal width for Setosa has a few outliers (dots outside the whiskers).

*   Virginica generally has the largest petal measurements.
